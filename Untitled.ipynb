##  Null all countries order the same total amount in dollars  Alternative some countries order larger amounts

```python
country = pd.DataFrame(c.execute("SELECT CompanyName,Country,SUM(UnitPrice*Quantity) AS TotalBought  FROM Customer JOIN [Order] ON Customer.ID = [Order].CustomerID JOIN OrderDetail ON [Order].ID=OrderDetail.OrderID GROUP BY Country ORDER BY TotalBought DESC ").fetchall())
sumcol1 = country[2].sum() #sum of all dollars or total dollars revenue 
country[3] = country[2] / sumcol1 #add the percentages
country[4] = round(country[3].cumsum(),2)#cumulative sum of Percentages
country
```

```python
top5 =country[2].nlargest(5)#top 5 revenues 
control = np.random.normal(country[2].mean(),country[2].std(),1000)
control =np.percentile(control,[75,81,87,93,99])
np.random.seed(10)
n_sim = 2000
power_sample = []
p = (np.empty(n_sim))
p.fill(np.nan)


for i in range(n_sim):
    control = np.random.normal(control.mean(),control.std(),len(control))
    experiment = np.random.normal(top5.mean(),top5.std(),len(top5))
    tstat = stats.ttest_ind(control,experiment)
    p[i] = tstat[1]
    num_null_rejects = len([x for x in p if x < .025])
    power = num_null_rejects/n_sim
power
```

This means that the best countries data is near normally distributed. There is a deviation from the expected normal distribution 7% of the time with 95% confidence that either the data is either too low (not enough of a top country) or too high ( the countries order more than they should given the distribution). That means that luckily the best countries are evenly balanced so that it would seem that the business can expect those countries to last because they fit into the normal distribution of normality. Alternative Hypothesis is true.

## Null top customers are normally distributed  Alternative top customers are not normal

```python
top5 =customer[5].nlargest(5)
np.random.seed(10)
n_sim = 2000
power_sample = []
p = (np.empty(n_sim))
p.fill(np.nan)
control = np.random.normal(customer[5].mean(),customer[5].std(),1000)
control =np.percentile(control,[75,81,87,93,99])

for i in range(n_sim):
    control = np.random.normal(control.mean(),control.std(),len(control))
    experiment = np.random.normal(top5.mean(),top5.std(),len(top5))
    tstat = stats.ttest_ind(control,experiment)
    p[i] = tstat[1]
    num_null_rejects = len([x for x in p if x < .025])
    power = num_null_rejects/n_sim
power
```

This test has such high power that I am sure that this data is not normally distributed. If it were normally distributed then the power would have been near 0. Since I would expect that given the population mean standard deviation and size to produce normally distributed numbers every time and since these normally distributed numbers did not match up to the actual numbers then the actual mean of the real population is not within the data and thus the data is not good. Or in other words the actual best customers do not actually represent what they should be as far as how much in total revenue they give to the company they are either too good of customers or not good enough of customers compared to the normal distribution. Alternative hypothesis is true.

## Null the discount does not affect quantity Alternative the discount affects the quantity

```python
#discount affect quantity
#how many average quantity no discount vs 5% vs 10%
customer_0 = pd.DataFrame(c.execute("SELECT ProductName,Discount,Quantity  FROM  OrderDetail JOIN Product ON OrderDetail.ProductID=Product.ID   ORDER BY ProductName DESC ").fetchall())
customer_0percent = customer_0.loc[customer_0[1] == 0,[0,2]]
customer_5percent = customer_0.loc[(customer_0[1] > 0) & (customer_0[1] <.051),[0,2]]
customer_10percent = customer_0.loc[(customer_0[1] > 0.05) & (customer_0[1] <.101),[0,2]]
customer_15percent = customer_0.loc[(customer_0[1] > 0.1) & (customer_0[1] <.151),[0,2]]
customer_20percent = customer_0.loc[(customer_0[1] > 0.15) & (customer_0[1] <.201),[0,2]]
customer_25percent = customer_0.loc[(customer_0[1] > 0.2) & (customer_0[1] <.251),[0,2]]
customer_5percent.groupby([0]).mean()
customer_0percent[2].mean()
customer_20percent[2].mean()
means = [customer_0percent[2].mean(),customer_5percent[2].mean(),customer_10percent[2].mean(),customer_15percent[2].mean(),customer_20percent[2].mean(),customer_25percent[2].mean()]
means
stds = [customer_0percent[2].std(),customer_5percent[2].std(),customer_10percent[2].std(),customer_15percent[2].std(),customer_20percent[2].std(),customer_25percent[2].std()]
stds
sample_sizes = [customer_0percent[2].count(),customer_5percent[2].count(),customer_10percent[2].count(),customer_15percent[2].count(),customer_20percent[2].count(),customer_25percent[2].count()]
sample_sizes
print(means,stds,sample_sizes)
```
